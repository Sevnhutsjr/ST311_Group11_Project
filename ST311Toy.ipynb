```python
cuda = torch.cuda.is_available()
cuda
```

```python
pipe = pipeline("text-generation", model="openai-community/gpt2-large", device=0 if cuda else -1)
```

```python
context = "An AI therapy bot hears a patient say in an angry manner- 'I feel like nobody listens to me'. It synthesises this information and provides the following useful actionable strategies that a therapist could advise such a patient to take. Here they are: "
answer=pipe(context, max_length=3000) #try a sample output to input.
```
